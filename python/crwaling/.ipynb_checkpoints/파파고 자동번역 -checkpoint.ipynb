```python
from selenium.webdriver.chrome.service import Service # 다운로드된 크롬드라이버 파일을 연결하기 위해 활용

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup 
import time
import pandas as pd

import warnings
warnings.filterwarnings("ignore") # 불필요한 Warning 메시지를 꺼줍니다.
```

```python
from selenium.webdriver.support.ui import WebDriverWait
```

```python
text="HI!"
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--incognito')
driver = webdriver.Chrome(options=chrome_options)

# 번역 페이지 열기
translate_url = 'https://papago.naver.com/?sk=en&tk=ko'  # sk=en: English & tk=ko: Korean
driver.get(translate_url)
time.sleep(5)

# id가 txtSource인 곳에 번역할 문장을 보냅니다.
driver.find_element(By.ID, 'txtSource').send_keys(text)
time.sleep(10)

# id가 txtTarget인 곳에서 텍스트를 가져옵니다.
translated_contents = driver.find_element(By.ID, 'txtTarget')
print(translated_contents.text)
time.sleep()

# 가상 브라우저 닫기
driver.close()
driver.quit()
```
