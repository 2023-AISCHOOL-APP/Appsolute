```python
import requests as req
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys # 입력을 위한 컴퓨터용 키보드
from selenium.webdriver.common.by import By # 선택자 선택의 기준
from bs4 import BeautifulSoup as bs
import time # 파일이나 폴더를 생성, 삭제해주는 라이브러리
import pandas as pd
import numpy as np
```

```python
title_list=[]
singer_list=[]
lyrics_list=[]
```

```python
# 계절 url번호
season_url=[492,254,781,1648,930,1825,386,5727]
# 봄,여름,가을,겨울,아침,점심,저녁,밤/새벽

for i in range(len(season_url)):
    url = f'https://music.bugs.co.kr/musicpd?tag_id={season_url[i]}'
    req.get(url) 

    driver= wb.Chrome()
    driver.get(url)
```

```python
url = 'https://music.bugs.co.kr/musicpd?order=list&tag_id=492&page=4'
req.get(url) 
# Response [406] : 브라우저 가 아닌코드로 접근한것을 차단함. 접근불가능
# 브라우저인척 접근하는법 (headers)
# 해더정보가 빈값으로 요청되었기 때문에 해더값을 포함해서 요청
driver= wb.Chrome()
driver.get(url)
soup1 =bs(driver.page_source, 'html.parser')
len_thumbnails = len(driver.find_elements(By.CLASS_NAME, 'thumbnails'))
for j in range(20,len_thumbnails):
    soup1 =bs(driver.page_source, 'html.parser')
    # div class="thumbnails" 안의 모든 a 태그 찾기
    thumbnails = driver.find_elements(By.CLASS_NAME, 'thumbnails')[j]
    thumbnails.click()
    soup2 = bs(driver.page_source, 'html.parser')
    len_track = len(driver.find_elements(By.XPATH, '//td[a[@class="trackInfo"]]'))
    for i in range(len_track):
        time.sleep(3)

        # 새로운 페이지의 HTML을 BeautifulSoup으로 파싱
        soup2 = bs(driver.page_source, 'html.parser')
        td_track_info = driver.find_elements(By.XPATH, '//td[a[@class="trackInfo"]]')[i]

        td_track_info.click()
        soup_track_info = bs(driver.page_source, 'html.parser')

        # 노래 제목 가져오기
        title_element = soup_track_info.select_one('.innerContainer h1')
        song_title = title_element.get_text(strip=True) if title_element else ''

        # 아티스트 가져오기
        artist_element = soup_track_info.select_one('tbody tr:first-child td a')
        artist_name = artist_element.get_text(strip=True) if artist_element else ''

        # 가사 가져오기
        lyrics_element = soup_track_info.find('xmp')
        lyrics_text = lyrics_element.get_text(strip=True) if lyrics_element else ''

        # 가져온 정보 출력
        title_list.append(song_title)
        singer_list.append(artist_name)
        lyrics_list.append(lyrics_text)

        time.sleep(3)
        driver.back()
    time.sleep(3)    
    driver.back()
```
